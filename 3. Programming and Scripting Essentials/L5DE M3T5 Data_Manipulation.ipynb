## **3.1 Finding and Fixing Missing Values**

Missing data is a common problem faced in real world datasets. The first step in tackling missing data is finding it. By inspecting missing data, we can analyse it and sometimes even determine if there's a systemic issue that's causing it.

For example, we could discover all customers from a certain region were missing height and weight data, and you might conclude the collection method for that region is different or broken.

````python
import pandas as pd

df_messy = pd.read_csv('https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/fitconstructor/user_data_messy.csv')
df_messy.head(5)
````

**Notice** how the 4th data entry in height shows up as **NaN**. NaN is how pandas represents missing values and it's short for Not A Number.

````python
# Show missing values as True or False, this can be confusing to  look at
df_messy.isna()
````

````python
# So let's display rows where any element is missing
df_messy[df_messy.isna().any(axis=1)]
````

- `df_messy.isna()` checks each element in the DataFrame for missing values and returns a DataFrame of the same shape, but with True for missing values and False for others.

- `.any(axis=1)` checks each row to see if it has any True values, which would mean that row has at least one missing value.

- `df_messy[...]` filters the original DataFrame to only include rows that meet the condition set inside the square brackets, i.e., rows with any missing values.

### **To fix these missing value we can employ 3 Strategies**

1. **Removing Rows**: You can use `dropna(`) to remove rows with missing values.
2. **Filling Values:** Use `fillna(`) to fill missing values with a specified value or using a specified method (e.g. replace with 0 or the median value)
3. **Forward Fill or Backward Fill:** Use the `fillna()` method with method parameter as '`ffill`' or '`bfill`'. This copies the previous or next rows' value into the entry with missing data.

#### **Removing rows with missing data**

````python
# This removes the rows with missing data from our dataframe
df_messy.dropna()
# we can use df_messy.dropna(inplace=True) to remove and automatically update df_messy
````

#### **Using fill_na() to replace with the mean**

**Note:**
- `fill_na(value`) can replace the missing data with the any `value`
- We can use the inplace argument to replace our data, for example:
  - `df_messy['height_cm'].fillna(mean_value, inplace=True)`

````python
# Compute the mean of the 'height_cm' column
mean_value = df_messy['height_cm'].mean()

# Fill missing values in the 'Age' column with the mean
df_messy['height_cm'].fillna(mean_value)
````

#### **We can also do foward and backward fills**

Notice how the 4th row under the height column takes the value of the previous above (164).

````python
df_messy.fillna(method='ffill')  # Forward fill
````

````python
df_messy.fillna(method='bfill')  # Backwards fill
````

### **4.1 Selecting data Using .loc[] and .iloc[]**

Pandas provides two essential methods for selecting data: `.loc[]` and` .iloc[]`. These methods are your keys to unlocking data within your DataFrame.

#### **The .loc[] method**

The `.loc[]` method allows you to select data based on its label or name, in both the row and column dimensions. In our example:

````python
import pandas as pd

# We can easily load CSV files with Pandas, either from a URL or a file stored on your system.
url_csv = 'https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/fitconstructor/user_data.csv'

# Load data from a CSV file
df = pd.read_csv(url_csv)

df.head()
````

````python
# Select the first row of the 'name' column
df.loc[0,'name']
````

Here, 0 refers to the row index and `'name'` refers to the column name. So when we use `.loc[0, 'name'`], we are specifically asking Pandas to give us the data that sits in the intersection of the first row `(index 0)` and the column named `'name'`.

In simpler terms, it's like saying, "Hey, Pandas! Could you please fetch me the name of the product that's listed in the first row of the DataFrame?"

The `loc` attribute allows you to select both rows and columns by labels. If you want to select all rows and some columns, you can do:

````python
selected_columns = df.loc[:, ['name', 'gender']]
selected_columns.head()
````

## **5.2 Pandas GroupBy**

Grouping is the process of **categorising** data based on one or more columns. It allows you to focus your analysis on specific sections of your data.

The `groupby` method in Pandas is a powerful tool for aggregating data into groups based on some criteria. It essentially involves splitting the data into different groups, applying a function to each group independently, and then combining the results.

**Basic Syntax**

The basic syntax for the groupby method is as follows:

- `DataFrame.groupby(by, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=<object>, observed=False)`

**Parameters:**

- **by**: Specifies the columns or a mapping you want to group by. You can also use a function or multiple arrays of the same length.

- **axis**: Determines the axis along which to group (0 for rows and 1 for columns). The default is 0.

- **level**: If the DataFrame has a MultiIndex, you can choose which level to group by.

- **as_index**: When True (the default), the keys you use to group will become an index in the new DataFrame. Set this to False if you want the keys to remain as columns.

- **sort**: Sorts the groups when set to True (the default).

- **group_keys**: Adds a hierarchical index when True, based on the group keys. Useful when you're using apply with the groups.

- **squeeze**: Squeezes the groups to a Series if they contain only one column.

- **observed**: Takes into account only observed categories for Categorical columns when set to True.


````python
df.groupby(by='gender').mean(numeric_only=True) # Use numeric_only = True, if you want to calculate the mean only for numeric columns
````

#### **Groupby with multiple columns**

````python
df.groupby(by=['gender','weight_kg'], sort=True).mean()
````

**Notice** the heirarchical nature of the output after performing a groupby.

When you perform a groupby operation followed by an aggregation function like sum, mean, etc., and you group by multiple columns, Pandas will return a DataFrame with a **MultiIndex** or **hierarchical index**. A hierarchical index allows you to have multiple levels of columns or rows, making it easier to work with higher dimensional data. However, they can be difficult to work with if you're not used to them.

**Why Hierarchical Indices?**

Hierarchical indices can be very useful when you need to summarize data in a structured and accessible way. For example, you might first group by "Country", and then by "State" within each country. The hierarchical index allows you to easily perform operations at different granularity levels.

**How to Flatten Hierarchical Index**

Use `df.reset_index()`



````python
df_grouped = df.groupby(by=['gender','weight_kg'], sort=True).mean().reset_index()
df_grouped
````

## **5.3 What is Aggregating?**

The `agg` function in Pandas allows you to perform multiple aggregate operations at once or perform different aggregate operations on different columns. The function is highly versatile and can accept a variety of input formats like strings, dictionaries, or even custom functions.

**Syntax: **

- `DataFrame.agg(func, axis=0, *args, **kwargs)`

- `func`: The function or functions to apply. Can be a single function or a list for applying multiple functions.
- `axis`: The axis to apply the function along (0 for applying function to each column).


````python
df.groupby('gender').agg({
    'height_cm': 'mean',
    'weight_kg': 'min'
})
````

**Differences Between agg and Basic groupby:**

- **Flexibility**: Using agg allows you to perform multiple and different kinds of operations at once, whereas a basic groupby followed by an aggregate function (e.g., mean(), sum()) performs the same operation on all selected columns.

- **Column Specification:** With agg, you can specify which columns to aggregate and how, making it more explicit and readable.

- **Multiple Functions**: agg can apply multiple functions to each column, resulting in a hierarchical index in the output DataFrame.

````python
# Using agg to specify operations for each column
df.groupby('gender').agg({'height_cm': 'mean', 'weight_kg': ['min', 'max']})
````

````python
# Applying multiple functions to 'height_cm' and 'weight_kg'
df.groupby('gender').agg({'height_cm': ['mean', 'max'], 'weight_kg': ['min', 'max']})
````

## **6.1 Method 1: Using merge() for SQL-like Join**
The merge() function provides a lot of flexibility and control over how you want to combine DataFrames. Here are the parameters we'll focus on:



```
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False)

```


Here are the some of the parameters we'll focus on:

- **left and right**: The DataFrames you want to merge.
- **how**: Type of merge to be performed. Options include 'left', 'right', 'outer', and 'inner'. The default is 'inner'.
- **on**: Column(s) that should be used to make the join. Must be found in both DataFrames.
- **left_on and right_on**: Columns from the left and right DataFrames to use as keys if the DataFrames do not have the same column for joining.
left_index and right_index: Use the index from the left/right DataFrame as the join key(s). If it is a MultiIndex, the number of keys in the other DataFrame's join keys must match.
- **sort**: Sort the result DataFrame by the join keys in lexicographical order. Defaults to False.



### **Now let's perform our first merge**

# Now let's perform our first merge

```
merged_data = pd.merge(user_data, crm_data, on='user_id')
```


This uses the '**inner**' join method by default and specifies **user_id** as the column common to both DataFrames for merging.

- **left**: This is user_data, the DataFrame that appears as the first argument in pd.merge(). It is one of the two DataFrames you are attempting to merge.

- **right**: This is crm_data, the DataFrame that appears as the second argument in pd.merge(). It is the other DataFrame you are merging with user_data.

- **on**: This specifies the column that both user_data and crm_data share, which in this case is 'user_id'. The merge function uses this column as the key for merging the two DataFrames.





````python
import pandas as pd

# We can easily load CSV files with Pandas, either from a URL or a file stored on your system.
user_data_csv = 'https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/fitconstructor/user_data.csv'
crm_data_csv = 'https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/fitconstructor/crm.csv'

# Load data from a CSV file
user_data = pd.read_csv(user_data_csv)
user_data.drop(columns = 'Unnamed: 10', inplace=True)

crm_data = pd.read_csv(crm_data_csv)
crm_data.head()
````

````python
# Now let's perform our first merge
merged_data = pd.merge(user_data, crm_data, on='user_id')
merged_data
````

#### **Observe the resulting dataframe merged_data now has all the user_data plus new columns from CRM which are number_of_visit	first_purchase_date	original_source**

- **merged_data:** This is the resulting DataFrame after the merge operation. It will contain all the columns from both user_data and crm_data, and rows will be combined based on the common user_id.

## **6.2 Method 2 - Understanding join() in Pandas**
The `join()` method in Pandas is used to combine rows of two or more data frames based on one or more keys. This method is similar to SQL joins. Here's a breakdown of the syntax and its key parameters:

**Syntax**:


```
DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
```

**Parameters:**

- **other:** The DataFrame you want to join with.
- **on:** Column or index level names to join on. These must be found in both DataFrames.
- **how:** Type of join to be executed (left, right, outer, inner). Default is left.
lsuffix and rsuffix: Suffixes to use for overlapping columns from the left and right DataFrames, respectively.
- **sort:** Sort the join keys lexicographically in the result DataFrame. Default is False.



````python
# Make sure 'user_id' is set as the index for both DataFrames
user_data.set_index('user_id', inplace=True)
crm_data.set_index('user_id', inplace=True)

# Perform the join
joined_data = user_data.join(crm_data)
joined_data
````

## **Method 3: Using concat() for Direct Concatenation**

The **concat()** function is used for stacking DataFrames either vertically or horizontally (technically not merging as not matching by a key is taking place). It is useful when you have DataFrames with similar structures and want to combine them into a single DataFrame.

To concatenate user_data and crm_data vertically:

````python
concatenated_data = pd.concat([user_data, crm_data], ignore_index=True)
concatenated_data
````
